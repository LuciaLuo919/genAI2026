```python
sns.kdeplot(sheight['GPA'])
plt.show()
```

```python
sample_10 = []

for i in range(200):
    sub_sample = sheight.sample(10, random_state=2045 + i)
    sub_mean = sub_sample['GPA'].mean()
    sample_10.append(sub_mean)
```

```python
sample_30 = []
for i in range(200):
    sub_sample = sheight.sample(30, random_state=2045 + i)
    sub_mean = sub_sample['GPA'].mean()
    sample_30.append(sub_mean)
```

```python
plt.figure(figsize=[6,6])
plt.subplots_adjust(hspace=0.8)

plt.subplot(311)
sns.kdeplot(sheight['GPA'], label='Sample')
plt.xlabel('GPA',fontsize=9)
plt.title('Distribution of GPA')

plt.subplot(312)
sns.kdeplot(sample_10, label='Sample')
plt.xlabel('mean',fontsize=9)
plt.title('Distribution of Sample Mean (sample size = 10)')

plt.subplot(313)
sns.kdeplot(sample_30, label='Sample')
plt.xlabel('mean',fontsize=9)
plt.title('Distribution of Sample Mean (sample size = 30)')
plt.show()
```
